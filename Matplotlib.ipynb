```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_data = np.random.rand(50)*100
y_data = np.random.rand(50)*100
```

```python
print(x_data)
print(y_data)
```

```python
#Scatter Plot (Plotted as Dots)
plt.scatter(x_data,y_data)
#plt.scatter(x_data,y_data , c ="#000" , marker="*")
plt.show()
```

```python
year = [2006+x for x in range(16)]
weights = [50+x for x in range(16)]
```

```python
#Line Plots
plt.plot(year,weights , lw=5)
plt.show()
```

```python
x=["a","b","c","d","e"]
y=[10,60,30,25,50]
```

```python
#Bar Chart
plt.bar(x,y , width =0.2)
plt.show()
plt.bar(x,y , width =0.2 , edgecolor="green", lw = 2)
plt.show()
```

```python
#Histogram (Frequency)
ages = np.random.normal(20,1.5,1000) # 20 ->mean , 1.5->Stand Dev , 1000->Size(No. of Samples)
plt.hist(ages,bins=20)
plt.show()
```

```python
#Pie Chart -> NotOverlapping
lang=["C#","C+","C++","Java"]
vote=[63,52,35,22]
plt.pie(vote , labels=lang, autopct="%1.1f%%" )
plt.show()
#use explode totake a piece out . Useful in Pesentation
```

```python
#Box Plot -> quartile , median , max,min , outliers
heights = np.random.normal(170,10,200)
plt.boxplot(heights)
plt.show()
```

```python
years = [2014,2015,2016,2017,2018,2019]
salary= [55000,66000,90000,150000,180000,220000]
plt.plot(years,salary)
income_ticks=[40000,80000,120000,160000,200000,230000]
plt.yticks(income_ticks)
plt.title("Salary Growth" , fontname = "FreeSerif")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

```python
stock_a=[25,53,23,54,34,43,76]
stock_b=[21,43,53,44,24,50,70]
plt.plot(stock_a , label="Stock A")
plt.plot(stock_b , label="Stock B")
plt.legend(loc="upper center")
plt.show()
```

```python
#We can use Multiple Styles for Pie Chart


from matplotlib import style
style.use("ggplot")
import numpy as np
```

```python
#Sub-plots:
x =np.arange(100)
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Cos")

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title("Random")

axs[1,1].plot(x,np.random.random(100))
axs[1,1].set_title("Random")
axs[1,1].set_xlabel("X-Axis")
axs[1,1].set_ylabel("Y-Axis")

fig.suptitle("Subplots")
plt.tight_layout() # Won't Overlap
plt.show()
#plt.savefig("four_plot.png") #Make File
```
